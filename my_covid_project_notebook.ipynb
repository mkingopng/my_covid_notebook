# 1.6: Summary of Findings

**The analysis concludes that:**

- Increased levels of vaccination have clearly reduced deaths due to COVID-19 to virtually zero.

- Increased levels of vaccination have generally been associated with reduced COVID-19 morbidity

- Increased levels of population fully vaccinated is a highly desirable and worthwhile public health goal.

- Governments are right to promote a public narrative that aggressively encourages mass inoculation.

**However:**

- The correlation between increased levels of vaccination and other COVID-19 morbidity variables has changed over the sample period.

- The emergence of COVID-19 variants like Delta in the sample population are a significant confounding variable, which has changed the relationship between increases in population fully vaccinated, COVID-19 R-0 and COVID-19 morbidity variables

- Changes in the relationship between increased levels of vaccination and most COVID-19 morbidity variables is closely correlated with inflection points in the emergence of Delta.

- Based on the UK experience, it is not reasonable to expect that once Australia reaches 70% 'life will return to normal'.

**Unfortunately there is growing evidence indicating that:** 

- 'Breakthrough' cases are occurring.

- Fully vaccinated people can and are being infected with COVID-19. 

- Fully vaccinated people who become infected with COVID-19 can and do get sick, requiring hospitalisation and ICU treatment.

- Fully vaccinated people can be asymptomatic carriers of COVID-19, thereby spreading the disease.

- That some variants are both more infectious and better equipped to 'break through' the protections offered by the current crop of vaccines.

- Newer variants will continue to emerge, and natural selection will favour new variants with resistance to existing vaccines. This will ultimately render the current crop of vaccines ineffective.

**Therefore:**

- the emerging consensus is that beyond the point at which 70% population fully vaccinated is achieved, non-pharmaceutical interventions (NPI's), vaccine boosters, and updated vaccines will continue to be required.

Merge aggregated with consolidated_df

## 2.6.5: Aggregate the data - step 4
Merge aggregated with variants_df

```python
aggregated = pd.merge(left=aggregated, right=variants_df, how='outer', on='date', sort=True, suffixes=("_x", "_y"))
aggregated['Beta'].interpolate(method='linear', inplace=True)
aggregated['Epsilon'].interpolate(method='linear', inplace=True)
aggregated['Gamma'].interpolate(method='linear', inplace=True)
aggregated['Iota'].interpolate(method='linear', inplace=True)
aggregated['Eta'].interpolate(method='linear', inplace=True)
aggregated['Delta'].interpolate(method='linear', inplace=True)
aggregated['Alpha'].interpolate(method='linear', inplace=True)
aggregated['non_who'].interpolate(method='linear', inplace=True)
aggregated['Lambda'].interpolate(method='linear', inplace=True)
# add a calculated column
aggregated['%_of_population_fully_vaccinated'] = (aggregated.people_fully_vaccinated / (aggregated.population * 0.79) * 100)
# drop the column named 'index'
aggregated = aggregated.drop(['index'], axis=1)
# change date to index
aggregated['date'] = pd.to_datetime(aggregated['date'])
# set the index
aggregated = aggregated.set_index('date')
```

## 4.1 Plot the total number of vaccinations over time

Because the COVID-19 vaccines are not approved for children (people <18 years old), the % of population fully vaccinated is conventionally reported as the % of the population >18 years old fully vaccinated. UK data indicates that 21% of the population is <18

```python
# Create figure and plot space
fig, ax = plt.subplots(figsize=(15, 7))

x = aggregated.index.values
y1 = aggregated['%_of_population_fully_vaccinated']
y2 = aggregated['daily_vaccinations']
ax2 = ax.twinx()

# Add x-axis and y-axis
ax.plot(x, y1, color = 'g', label = '%_of_population_fully_vaccinated')
ax2.plot(x, y2, color = 'b', label = 'daily_vaccinations')

# Set title and labels for axes
ax.set_xlabel('Date', fontdict=axesdict)
ax.set_ylabel('% of Population Fully Vaccinated', fontdict=axesdict)
ax2.set_ylabel('Daily Vaccinations', fontdict=axesdict)
ax.set_title('% of Population Fully Vaccinated over Time', fontdict=titledict)

# Define the date format
date_form = DateFormatter("%m-%y")
ax.xaxis.set_major_formatter(date_form)

# set v lines for important dates
ax.axvline( dt.datetime(2021, 5, 3), label=label_1, c='r', linestyle=vlinestyle)
ax.axvline(dt.datetime(2021, 6, 1), label=label_2, c='b', linestyle=vlinestyle)
ax.axvline(dt.datetime(2021, 7, 23), label=label_3, c='black', linestyle=vlinestyle)

# daily_vaccinations definition: new doses administered per day (7-day smoothed).
ax.legend(loc='center left', bbox_to_anchor=(0.25, -0.25))
ax2.legend(loc='center right', bbox_to_anchor=(0.25, -0.25))
plt.savefig('total_vaccinations.png')
plt.show()
```

### Comments
Daily vaccinations, depicted by the blue line, increase rapidly from January to mid-March 2021. Daily vaccination rates fluctuated between 400,000 and 600,000 per day from mid-March to mid-June. Daily vaccination rates have slowed thereafter.
